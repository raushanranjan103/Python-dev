```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
y1=x**2
y2=2*x+3
x,y1,y2
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
plt.plot(x,y1,color='red',label="kiwi",marker='o')
plt.plot(x,y2,color='green',label="apple",linestyle="dashed")
plt.xlabel("time")
plt.ylabel("price")
plt.title("price of fruits over time")
plt.show()
```
