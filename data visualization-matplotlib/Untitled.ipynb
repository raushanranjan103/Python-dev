```python


x=np.arange(10)
y1=x**2
y2=2*x+3
x,y1,y2
plt.figure(figsize=(10,10))
plt.style.use("seaborn")
plt.scatter(x,y1,color='red',label="kiwi",marker='<')
plt.scatter(x,y2,color='green',label="apple",linestyle="dashed")
plt.xlabel("time")
plt.ylabel("price")
plt.title("price of fruits over time")
plt.show()
plt.show()
```
